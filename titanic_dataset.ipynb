```python
import pandas as pd


df = pd.read_csv('Titanic-Dataset.csv')

df.head()
```

```python

survivors = df[df['Survived'] == 1]   # Filter for survivors


survival_list = survivors[['PassengerId', 'Name','Sex']]


survival_list.head()  # Display the survival list
```

```python
import matplotlib.pyplot as plt


survival_counts = df['Survived'].value_counts()  # Count the number of survivors and non-survivors

# Create a bar plot
plt.bar(survival_counts.index, survival_counts.values, color=['red', 'green'])


plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks([0, 1], ['Did not Survive', 'Survived'])


plt.show()
```

```python
male_survivors = df[(df['Survived'] == 1) & (df['Sex'] == 'male')]   # Filter for male survivors


female_survivors = df[(df['Survived'] == 1) & (df['Sex'] == 'female')]   # Filter for female survivors
print(f"Number of male survivors: {len(male_survivors)}")
print(f"Number of female survivors: {len(female_survivors)}")
```

```python

survival_list.to_csv('survival_list.csv', index=False)   # Save the survival list to a file

# Download the file
from google.colab import files
files.download('survival_list.csv')

```
